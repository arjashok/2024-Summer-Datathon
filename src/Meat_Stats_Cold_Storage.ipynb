```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Meat_Stats_Cold_Storage - Red Meat (Beef vs. Veal vs. Pork vs. Lamb and Mutton)
df = pd.read_csv('../datasets/Meat_Stats_Cold_Storage.csv')

# converts 'Year' and 'Month' columns to datetime object
df['Date'] = pd.to_datetime(df['Year'].astype(str) + '-' + df['Month'].astype(str), format='%Y-%m')

# Beef
beef_condition = (df['Animal'] == 'Beef')
beef_df = df[beef_condition].copy()
beef_df['Weight'] = beef_df['Weight'].astype(float)
# summing all weights from a single year into one data value so that the plot will be more readable
beef_df = beef_df.groupby(beef_df['Date'].dt.year)['Weight'].sum().reset_index()
plt.plot(beef_df['Date'], beef_df['Weight'], label='Beef')

# Veal
veal_condition = (df['Animal'] == 'Veal')
veal_df = df[veal_condition].copy()
veal_df['Weight'] = veal_df['Weight'].astype(float)
# summing all weights from a single year into one data value so that the plot will be more readable
veal_df = veal_df.groupby(veal_df['Date'].dt.year)['Weight'].sum().reset_index()
plt.plot(veal_df['Date'], veal_df['Weight'], label='Veal')

# Pork
pork_condition = (df['Animal'] == 'Pork')
pork_df = df[pork_condition].copy()
pork_df['Weight'] = pork_df['Weight'].astype(float)
# summing all weights from a single year into one data value so that the plot will be more readable
pork_df = pork_df.groupby(pork_df['Date'].dt.year)['Weight'].sum().reset_index()
plt.plot(pork_df['Date'], pork_df['Weight'], label='Pork')

# Lamb and Mutton
lamb_condition = (df['Animal'] == 'Lamb and Mutton')
lamb_df = df[lamb_condition].copy()
lamb_df['Weight'] = lamb_df['Weight'].astype(float)
# summing all weights from a single year into one data value so that the plot will be more readable
lamb_df = lamb_df.groupby(lamb_df['Date'].dt.year)['Weight'].sum().reset_index()
plt.plot(lamb_df['Date'], lamb_df['Weight'], label='Lamb and Mutton')

plt.xlabel('Date')
plt.ylabel('Weight (Million Pounds)')
plt.yticks(range(0, 8000, 500)) # set y-axis ticks at intervals
plt.title('Amount of Red Meat in Cold Storage')
plt.legend()
plt.savefig('Red Meat Cold Storage.png')
plt.show()
```

```python
# Meat_Stats_Cold_Storage - Poultry (Broiler vs. Other Chicken vs. Turkey vs. Frozen Eggs)
df = pd.read_csv('../datasets/Meat_Stats_Cold_Storage.csv')

# converts 'Year' and 'Month' columns to datetime object
df['Date'] = pd.to_datetime(df['Year'].astype(str) + '-' + df['Month'].astype(str), format='%Y-%m')

# Broiler
# No data on broilers before May-2003
broiler_condition = (df['Animal'] == 'Broiler')
broiler_df = df[broiler_condition].copy()
broiler_df['Weight'] = broiler_df['Weight'].astype(float)
# summing all weights from a single year into one data value so that the plot will be more readable
broiler_df = broiler_df.groupby(broiler_df['Date'].dt.year)['Weight'].sum().reset_index()
plt.plot(broiler_df['Date'], broiler_df['Weight'], label='Broiler')

# Other Chicken
other_chicken_condition = (df['Animal'] == 'Other Chicken')
other_chicken_df = df[other_chicken_condition].copy()
other_chicken_df['Weight'] = other_chicken_df['Weight'].astype(float)
# summing all weights from a single year into one data value so that the plot will be more readable
other_chicken_df = other_chicken_df.groupby(other_chicken_df['Date'].dt.year)['Weight'].sum().reset_index()
plt.plot(other_chicken_df['Date'], other_chicken_df['Weight'], label='Other Chicken')

# Turkey
turkey_condition = (df['Animal'] == 'Turkey')
turkey_df = df[turkey_condition].copy()
turkey_df['Weight'] = turkey_df['Weight'].astype(float)
# summing all weights from a single year into one data value so that the plot will be more readable
turkey_df = turkey_df.groupby(turkey_df['Date'].dt.year)['Weight'].sum().reset_index()
plt.plot(turkey_df['Date'], turkey_df['Weight'], label='Turkey')

# Frozen Eggs
eggs_condition = (df['Animal'] == 'Frozen Eggs')
eggs_df = df[eggs_condition].copy()
eggs_df['Weight'] = eggs_df['Weight'].astype(float)
# summing all weights from a single year into one data value so that the plot will be more readable
eggs_df = eggs_df.groupby(eggs_df['Date'].dt.year)['Weight'].sum().reset_index()
plt.plot(eggs_df['Date'], eggs_df['Weight'], label='Frozen Eggs')

plt.xlabel('Date')
plt.ylabel('Weight (Million Pounds)')
plt.yticks(range(0, 11000, 1000)) # set y-axis ticks at intervals
plt.title('Amount of Poultry in Cold Storage')
plt.legend()
plt.savefig('Poultry Cold Storage.png')
plt.show()
```

```python
# Meat_Stats_Cold_Storage - Red Meat vs. Poultry
df = pd.read_csv('../datasets/Meat_Stats_Cold_Storage.csv')

# converts 'Year' and 'Month' columns to datetime object
df['Date'] = pd.to_datetime(df['Year'].astype(str) + '-' + df['Month'].astype(str), format='%Y-%m')

# Red Meat
red_meat_condition = (df['Type_Of_Meat'] == 'Red Meat')
red_meat_df = df[red_meat_condition].copy()
red_meat_df['Weight'] = red_meat_df['Weight'].astype(float)
# summing all weights from a single year into one data value so that the plot will be more readable
red_meat_df = red_meat_df.groupby(red_meat_df['Date'].dt.year)['Weight'].sum().reset_index()
plt.plot(red_meat_df['Date'], red_meat_df['Weight'], label='Red Meat')

# Poultry
poultry_condition = (df['Type_Of_Meat'] == 'Poultry')
poultry_df = df[poultry_condition].copy()
poultry_df['Weight'] = poultry_df['Weight'].astype(float)
# summing all weights from a single year into one data value so that the plot will be more readable
poultry_df = poultry_df.groupby(poultry_df['Date'].dt.year)['Weight'].sum().reset_index()
plt.plot(poultry_df['Date'], poultry_df['Weight'], label='Poultry')

plt.xlabel('Date')
plt.ylabel('Weight (Million Pounds)')
plt.yticks(range(0, 17000, 2000)) # set y-axis ticks at intervals
plt.title('Red Meat vs. Poultry in Cold Storage')
plt.legend()
plt.savefig('Red Meat vs. Poultry Cold Storage.png')
plt.show()
```
